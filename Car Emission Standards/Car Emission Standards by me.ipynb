```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t, norm
plt.rcParams['figure.figsize'] = 10,6
```

```python
small_data = np.loadtxt('emission_small.txt')
large_data = np.loadtxt('emission_large.txt')
standard = 1
```

```python
plt.hist(small_data, bins=200, label='Small')
plt.hist(large_data, bins=20, histtype='step', label='Large')
plt.axvline(standard, label='Standard')
plt.legend(loc=2);
```

```python
small_data.shape
```

```python
large_data.shape
```

```python
mean, std = small_data.mean(), small_data.std() / np.sqrt(small_data.size)
xs = np.linspace(0.5, 1.4, 200)
ps = t.pdf(xs, 9, mean, std)   #student's t RV function is used when there are less number of samples
prob_fail = t.sf(standard, 9, mean, std)
print(f"There is {100 * prob_fail:.2f}% chance that vehicle will fail the emission test.")
```

```python
plt.hist(small_data, bins=200, label='Small')
plt.plot(xs, ps, label='Confidence of Mean')
plt.axvline(standard, ls=':', label='Standard')
plt.fill_between(xs, ps, where=xs >= standard, alpha=0.2, color='r', label='P(Fail)')
plt.legend(loc=2);
```

```python
mean, std = large_data.mean(), large_data.std() / np.sqrt(large_data.size)
xs = np.linspace(0.5, 1.4, 200)
ps = norm.pdf(xs, mean, std)   #normal RV function is used since there is large number of samples
prob_pass = norm.cdf(standard, mean, std)
print(f"There is {100 * prob_pass:.2f}% chance that vehicle will pass the emission test.")

plt.hist(large_data, bins=15, histtype='step', density=True, label='Large')
plt.plot(xs, ps, label='Confidence of Mean')
plt.plot(xs, norm.pdf(xs, mean, large_data.std()), label='Dist of Observables')
plt.axvline(standard, ls=':', label='Standard')
plt.fill_between(xs, ps, where=xs >= standard, alpha=0.2, color='r', label='P(Fail)')
plt.legend(loc=2)
plt.xlabel('Emission');
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
```

```python
data = np.loadtxt('dataset.txt')
```

## Variance

```python
def get_variance(xs):
    mean = np.mean(xs)
    summed = 0
    for x in xs:
        summed += (x - mean)**2
    return summed / (len(xs) - 1)
```

```python
get_variance([1,2,3,4,5])
```

```python
get_variance(data)
```

```python
np.var(data)
```

## Skewness - 
Skewness is a measure of symmetry, or more precisely, the lack of symmetry. A distribution, or data set, is symmetric if it looks the same to the left and right of the center point. 

```python
def get_skewness(ab):
    mean = np.mean(ab)
    var = np.var(ab)
    summed = 0
    for x in ab:
        summed += (x - mean)**3
    return (summed / (len(ab))) / (var**1.5)
```

```python
get_skewness([1,2,3,4,5])
```

```python
skewness = st.skew(data)
print(skewness, get_skewness(data))
```

```python
xs = np.linspace(data.min(), data.max(), 100)
ys1 = st.norm.pdf(xs, loc=mean, scale=std)
ys2 = st.skewnorm.pdf(xs, skewness, loc=mean, scale=std)

plt.hist(data, bins=50, density=True, histtype='step', label='Data')
plt.plot(xs, ys1, label='Normal Approximation')
plt.plot(xs, ys2, label='Skewnormal Approximation')
plt.ylabel('Probability')
plt.legend();
```

```python
xs = np.linspace(data.min(), data.max(), 100)
ys1 = st.norm.pdf(xs, loc=mean, scale=std)
ps = st.skewnorm.fit(data)   #return skewness, mean, std
ys2 = st.skewnorm.pdf(xs, *ps)   #this will give us the pdf with mean adjusted for the skewness

plt.hist(data, bins=50, density=True, histtype='step', label='Data')
plt.plot(xs, ys1, label='Normal Approximation')
plt.plot(xs, ys2, label='Skewnormal Approximation')
plt.ylabel('Probability')
plt.legend();
```

## Kurtosis
Kurtosis is a measure of whether the data are heavy-tailed or light-tailed relative to a normal distribution.

```python
def get_kurtosis(ab):
    mean = np.mean(ab)
    var = np.var(ab)
    summed = 0
    for x in ab:
        summed += (x - mean)**4
    return (summed / (len(ab))) / (var**2)
```

```python
get_kurtosis([1,2,3,4,5])
```

```python
kurtosis = st.kurtosis(data)
print(kurtosis, get_kurtosis(data))
```

```python
kurtosis = st.kurtosis(data, fisher=False)
print(kurtosis, get_kurtosis(data))
```

## Percentiles

```python
pc = np.linspace(0, 100, 10)
x_p = np.percentile(data, pc)

xs = np.sort(data)
ys = np.linspace(0, 1, len(data))

plt.plot(xs, ys*100, label='ECDF')
plt.plot(x_p, pc, label='Percentiles', marker='.', ms=10)
plt.ylabel('Percentile')
plt.legend();
```

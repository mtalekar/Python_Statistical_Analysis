```python
import scipy.stats as st
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5,10,200)
ks = np.arange(50)
```

## Discrete PMFs (Probability Mass Functions)
Binomial Distribution, Poisson distribution

```python
plt.rcParams['figure.figsize'] = 10,6
```

```python
pmf_binom = st.binom.pmf(ks, 50, 0.25)
plt.bar(ks, pmf_binom, label='Binomial Example (Dice)', alpha=0.8)

pmf_poisson = st.poisson.pmf(ks, 30)    #30 is like the right characteristic, the right/actual/true value
plt.bar(ks, pmf_poisson, label='Poisson Example (Car Crash)', alpha=0.8)
plt.legend()
```

```python
#pmf of geting 10 outcomes of 50 events, each having probability of 0.25
print(st.binom.pmf(10, 50, 0.25))

#one night 30 car crashs, what is the probabililty of getting 50
print(st.poisson.pmf(50,30))
```

## Continous Distributions
Uniform, Normal, Exponential, student-t, log-normal, skew-normal.

```python
pdf_uniform = st.uniform.pdf(xs, -4, 10)  #lower and upper bound
plt.plot(xs, pdf_uniform, label="Uniform(-4,6)")

pdf_normal = st.norm.pdf(xs, loc=5, scale=2)
plt.plot(xs, pdf_normal, label="Normal(5, 2)")

pdf_exponential = st.expon.pdf(xs, loc=-2, scale=2)
plt.plot(xs, pdf_exponential, label="Exponential(0.5)")

pdf_studentt = st.t.pdf(xs, 1)   #degree of freedom, which is genrally n-1
plt.plot(xs, pdf_studentt, label="Student-t(1)")

pdf_lognorm = st.lognorm.pdf(xs, 1)    #s parameter, telling how skewed the pdf is
plt.plot(xs, pdf_lognorm, label="Lognorm(1)")

pdf_skewnorm = st.skewnorm.pdf(xs, -6)
plt.plot(xs, pdf_skewnorm, label="Skewnorm(-6)")

plt.legend()
plt.ylabel("Prob")
plt.xlabel("x");
```

```python
plt.plot(xs, st.t.pdf(xs, 1, loc=4, scale=2), label='In built')
plt.plot(xs, st.t.pdf((xs-4)/2, 1, loc=0, scale=1), label='Manually')   #change in xs, makes the changes equivalent to loc and scale
plt.legend()
```

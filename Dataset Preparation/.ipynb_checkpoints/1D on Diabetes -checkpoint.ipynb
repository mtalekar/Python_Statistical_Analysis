```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
d1 = pd.read_csv('clean_diabetes.csv')
```

```python
d1.head()
```

# Histograms Plots

```python
plt.hist(d1.Glucose, label='Glucose', bins=15)
plt.hist(d1.BMI, label='BMI', bins=15)
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
bins1 = np.linspace(min(d1.Glucose.min(),d1.BMI.min()), max(d1.Glucose.max(),d1.BMI.max()), 50)
plt.hist(d1.Glucose, label='Glucose', bins=bins1)
plt.hist(d1.BMI, label='BMI', bins=bins1)
plt.ylabel('Counts')
plt.legend()
plt.show()
```

```python
bins1 = np.linspace(min(d1.Glucose.min(),d1.BMI.min()), max(d1.Glucose.max(),d1.BMI.max()), 50)
plt.hist(d1.Glucose, label='Glucose', bins=bins1, density=True)
plt.hist(d1.BMI, label='BMI', bins=bins1, density=True)
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
bins1 = np.linspace(min(d1.Glucose.min(),d1.BMI.min()), max(d1.Glucose.max(),d1.BMI.max()), 50)
plt.hist(d1.Glucose, label='Glucose', bins=bins1, density=True, histtype='step', lw=3)
plt.hist(d1.BMI, label='BMI', bins=bins1, density=True, histtype='step', ls=':')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
plt.rcParams['figure.figsize'] = 12,6
```

```python
bins1 = np.linspace(min(d1.Glucose.min(),d1.BMI.min()), max(d1.Glucose.max(),d1.BMI.max()), 50)
plt.hist([d1.Glucose, d1.BMI], label='Stacked', bins=bins1, density=True, histtype='barstacked', alpha = 0.5)
plt.hist(d1.Glucose, label='Glucose', bins=bins1, density=True, histtype='step', lw=3)
plt.hist(d1.BMI, label='BMI', bins=bins1, density=True, histtype='step', ls=':')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

```python
bins1 = 50
plt.hist([d1.Glucose, d1.BMI], label='Stacked', bins=bins1, density=True, histtype='barstacked', alpha = 0.5)
plt.hist(d1.Glucose, label='Glucose', bins=bins1, density=True, histtype='step', lw=3)
plt.hist(d1.BMI, label='BMI', bins=bins1, density=True, histtype='step', ls=':')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

---
Empirical Cumulative Distribution Function

```python
sd1 = np.sort(d1.BMI)
sd2 = np.sort(d1.Glucose)
cdf = np.linspace(0,1,752)
```

```python
plt.plot(sd1, cdf, label='BMI CDF')
plt.hist(d1.BMI,histtype='step', density=True, alpha=0.5)
plt.plot(sd2, cdf, label='BMI Glucose')
plt.hist(d1.Glucose,histtype='step', density=True, alpha=0.5)
plt.legend()
```
